### Task_1

Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
# Исходя из условий, применима формула Бернулли.
C = factorial(100) / ( factorial(85) * factorial(100-85))
P = C * 0.8**85 * 0.2**15
P
```

### Task_2

Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

```python
# Исходя из условий, применима формула Пуассона.
Л = 5000 * 0.0004
# Вероятность, что ни одни не перегорит (перегорит 0):
P0 = Л**0 * math.e**-Л / factorial(0)
# Вероятность, что 2 перегорят:
P2 = Л**2 * math.e**-Л / factorial(2)
P0, P2
```

### Task_4:

В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

#### a) Все мячи белые:

```python
# Через условную вероятность:
# Для первого ящика вероятность что оба белые:
P1 = 7/10 * 6/9
print('В первой 2 белых:', P1)

# Для второго ящика вероятность что оба белые:
P2 = 9/11 * 8/10
print('Во второй 2 белых:', P2)

P = P1 * P2
'Ответ:', P
```

#### б) Ровно 2 мяча белые:

```python
# 2 мяча белые это варианты: 1. в первой 2 белый, во второй 0, 2. В первой 1 белый, во второй 1 белый, 3. В первой 0 белых, во
# второй 2 белых.

# 1)
# В первой 2 белых - P1 из предыдущей подзадачи (а).
# Во второй 0 белых: в корзине 2 не белых, соответственно одно сочетание.
# Общее кол-во сочетаний 2 из 11 во второй корзине:
C = factorial(11) / ( factorial(2) * factorial(11-2) )
P2корз = 1 / C

P_1 = P1 * P2корз
P_1

# 2)
# В первой корзине:
# 7 белых, по 1 это 7 комбинаций - 7 удовлетворяющих исходов и 3 комбинации из не белых, итого 21.
# Во второй корзине 9 белых - это 9 удовлетворяющих комбинаций из 1 и 2 из не белых, итого 18.

# Общее кол-во исходов для корзины 1:
C1 = factorial(10) / ( factorial(2) * (factorial(10-2)) )
# Общее кол-во исходов для корзины 2:
C2 = factorial(11) / ( factorial(2) * (factorial(11-2)) )

print('Из второй 1 белый:', 18/C2)
print('Из первой 1 белый:', 21/C1)

P_2 = 21/C1 * 18/C2

# 3)
# Удовл. в первой корзине, 3 не белых, 2 из 3 это 3 комбинации.
# Общее кол-во сочетаний посчитали в (2), это C1.
P1 = 3 / C1
# Удовл. во второй корзине: 2 белых, 2 из 2 = 1 комбинация.
# Общее кол-во сочетаний во 2-й: посчитано - C2 из варианта (2):
P2 = 1 / C2
P_3 = P1 * P2

# Удовлетворяет любой из этих 3-х вариантов, т.е. логическое ИЛИ
P = P_1 + P_2 + P_3
'Ответ:', P
```

#### в) Какова вероятность того, что хотя бы один мяч белый?

```python
# Хотя бы 1 мяч белый - или 1 мяч белый или 2 мяча белые или 3 или 4.
# В предыдущих подзадачах посчитали для 2 и 4, осталось через ИЛИ добавить для 1 и 3.
```

```python
# Для 1 мяча: (из первой берем один белый, из второй ни одного белого) или (из второй берем 1 белый, из первой - ни одного):

# 1 белый из 1-й корзины.
P1 = 21 / ( factorial(10) / ( factorial(2) * (factorial(10-2)) ) )
# 0 белых из 2-й корзины:
P0 = 1 / ( factorial(11) / ( factorial(2) * (factorial(11-2)) ) )

P = P1 + P0

# 0 белых из 1-й корзины (берем из предыдущих расчетов):
P1 = 3 / (factorial(10) / ( factorial(2) * (factorial(10-2)) ))

# 1 белый во второй корзине (берем из предыдущих расчетов):
P0 = 18 / (factorial(11) / ( factorial(2) * (factorial(11-2)) ))
P_ = P1 * P0

P = P + P_
print('Вероятность достать ровно 1 мяч:', P)
```

```python
# Для 3 мячей. Это варианты: (из первой 2 белых и из второй 1 белый) ИЛИ (из первой 1 белый, из второй 2 белых).
# Берем вычисления из предыдущих подзадач и стадий.
P = 0.46666666666666656 * 0.32727272727272727 + 0.4666666666666667 * 0.6545454545454545
P
```
